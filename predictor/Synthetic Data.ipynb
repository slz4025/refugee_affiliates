```python
import numpy as np
import pandas as pd
```

```python
aff_cities = pd.read_csv('./Affiliate-City-to-Id2.csv', header=None)
```

```python
# predictors and outcome used by Bansak paper, along with location_id
# most predictors are binary, with 1 the positive signal
# gender: M is 0, F is 1
# geoid: US Census place geoid for resettlement location
# employed: 90-day employment outcomes
df = pd.DataFrame(columns=[
    'english_speaking', 'gender', 'arrival_age', 'educ_high',
    'arrival_year', 'arrival_month', 'geoid', 'employed'
])

# total number of refugees in dataset
n_refugees = 10000

avg_english_speaking_rate = 0.56
# working-age refugees
age_low = 18
age_high = 64
# percent of adult refugees with a tertiary education, from Science paper
avg_educ_high_rate = 0.1057
# arrival year range
arrival_year_low = 2011
arrival_year_high = 2016
# average employment rate under actual assignment, from Science paper
avg_employment_rate = 0.34

for i in range(n_refugees):
    english_speaking = np.random.choice([0,1], p=[1-avg_english_speaking_rate, avg_english_speaking_rate])
    gender = np.random.choice([0,1], p=[0.5, 0.5])
    arrival_age = np.random.choice(range(age_low, age_high+1))
    educ_high = np.random.choice([0,1], p=[1-avg_educ_high_rate, avg_educ_high_rate])
    arrival_year = np.random.choice(range(arrival_year_low, arrival_year_high+1))
    arrival_month = np.random.choice(range(1, 13))
    employed = np.random.choice([0,1], p=[1-avg_employment_rate, avg_employment_rate])
    geoid = np.random.choice(aff_cities.iloc[:,1])
    df.loc[i] = [english_speaking, gender, arrival_age, educ_high, 
                 arrival_year, arrival_month, geoid, employed]
    
print(df)    
```

```python
df.to_csv('./synth-refugees-10000.csv')
```
